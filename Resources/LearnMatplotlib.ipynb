## Subplots and Legends

```python
days = [1,2,3,4,5]
original_price = [10,20,25,30,40]
predicted_price = [8,21,25,28,38]
```

```python
plt.plot(days, original_price, 'b-o', label='original price');
plt.plot(days, predicted_price, 'r-o', label='predicted price');
plt.legend()
plt.title('Plots with legends');
plt.xlabel('weekdays');
plt.ylabel('Price');
```

## subplots

```python
#First subplots - Original Price
plt.subplot(3,1,1);
plt.plot(days, original_price, 'b-o', label='original price');
plt.title('Original price over time');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.legend()

#Second subplots - Predicted Price
plt.subplot(3,1,2);
plt.plot(days, predicted_price, 'r-o', label='predicted price');
plt.title('Predicted price over time');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.legend()

#Third subplots - Difference between Original and Predicted Price
price_difference = [o-p for o,p in zip(original_price,predicted_price)]
plt.subplot(3,1,3);
plt.bar(days, price_difference, color='purple', label='Price Difference');
plt.title('Difference between Original and Predicted Price');
plt.xlabel('days');
plt.ylabel('Price');
plt.legend()

plt.tight_layout()
```

## Saving the Figure

```python
plt.plot(days, original_price, 'b-o', label='original price');
plt.plot(days, predicted_price, 'r-o', label='predicted price');
plt.legend()
plt.title('Plots with legends');
plt.xlabel('weekdays');
plt.ylabel('Price');
plt.savefig('example_predictions.png', dpi=300)#save the image
```
